# Embodiment Questionnaire Analysis

This notebook provides an starting point to use Principal Component Analysis (PCA) to study the responses of 
the Embodiment questionnaire proposed by Gonzalez-Franco & Peck.

```r
library(reshape)
library(ggplot2)
library(psych)

#library(RCurl)
#URL <- "https://notebooks.azure.com/margon/libraries/EmbodimentQuestionnairePCA/html/responses.csv"
#x <- getURL(URL)
#d <- read.csv(x)
#d
```

If they are uploaded locally with the notebook.
In this example we have generated a random csv that simulates a within subjects study with two conditions.

```r
d <- read.csv("responses.csv")
d
```

For the factor analysis we remove all variables that are not questionnaire responses.
In this example we remove the first two variables which are the subjectID and the condition.

```r
q_embodiment<-d[,-(1:2)] 
```

We check if the questions are correlated to eachother.

If variables are not correlated (i.e., if the variables were all orthogonal), there would be no way to combine the variables as factors or components. One basic test is Bartlett’s test of sphericity. The null hypothesis of the test is that the correlation matrix is an identity matrix– or that the matrix has one’s on the diagonal and zeroes on all the off diagonals. 

The test statistic follows a chi square distribution and to proceed, we would want to see statistically significant results (p<0.05) in order to proceed with a factor analysis.


```r
corMatrix<-cor(q_embodiment)
vars<-ncol(q_embodiment)

library(psych)
cortest.bartlett(corMatrix,n=nrow(q_embodiment))
```

Since the results indicate that the p value is < .05 (statistically significant) the PCA can be done. 

To further test if the factor analysis is adecuate we also check for KMO values over .5, in this case the combine value is 0.61.
This test measures the sampling adequacy for each variable in the model and the complete model. 

```r
KMO(corMatrix)
```

If some values are below 0.5 those question are better discarded from the analysis. In our case q1, q13, q14, q20 and q24 could be eliminated

```r
excluded <- names(q_embodiment) %in% c("q1","q13", "q14", "q20","q24") 
q_embodiment<-q_embodiment[!excluded]
```

We run a Principal Component Analysis

```r
# principal component analysis
res<-principal(q_embodiment , nfactors=ncol(q_embodiment), rotate="none")
res
plot(res$values, type="b")
```

We check for eigenvalues over 1 (represented by the variable SS Loadings).

In this example there are 5. We can also observe the curve in the plot reduces the slope.

We recalculate the PCA with the selected number of factors. We use obliminal rotation to optimize the algerba orientation.

Finally, we retrieve the loadings for each component. I.e. how important are the different questions in the questionnaire to describe each component/factor.

```r
nfactors<-5

res<-principal(q_embodiment , nfactors=nfactors, rotate="oblimin", scores=TRUE)
res$loadings
```

Ideally in total the cummulative explanation should be greater than 0.5, to show that most of the Variance in the data can be explained by these factors.

It is also interesting to try to interpret the factors in your analyis.
For example TC1 could be representing Body Ownership, (by the importance of the questions that load it), while TC2 could be more explanatory of Agency.

Finally. You would recalculate the factors per participant using the loadings and the original questionnaire

```r
q_f <- d

for (i in 1:nfactors){
    f<-res$loadings[,i]
    q<-rowMeans(q_embodiment*matrix( rep(f, nrow(q_embodiment)), 
                                    nrow=nrow(q_embodiment), byrow=TRUE))
    #print(q)
    q_f<-cbind(q_f,q)
    colnames(q_f)[ncol(q_f)] <- paste("f",i, sep="")
}
q_f
```

Plot the factors per condition.

```r
library(reshape)
factors_long<- melt(q_f, id.vars = c("subjectID", "condition"), measure.vars = paste('f', 1:nfactors, sep=""))
#factors_long

ggplot(factors_long, aes(x=variable, y=value, fill=factor(condition))) +    
    geom_boxplot() + ylim(-3,3)
```

Non-parametric Statistical Test. This can be done for each factor comparing each condition.
Depending on the experimental design you will used paired comparisons if it is within subjects (all participants completed both conditions) or not paired, for between subjects design.

```r
for (i in 1:nfactors){
    selected <- q_f[c("condition", paste("f",i, sep=""))]
    print(paste("f",i, sep=""))
    names(selected)<-c("condition", "factor")
    print(wilcox.test(factor~condition, paired = TRUE, conf.int = TRUE, exact=FALSE, data=selected))
    
    #if more than 2 conditions use friedman
    selected <- q_f[c("condition", "subjectID", paste("f",i, sep=""))]
    names(selected)<-c("condition", "subjectID", "factor")
    results <- friedman.test(factor ~ condition | subjectID, data=selected)
    print(results)
}
```

In this case (with random data), there are no significant differences between the conditions for any of the factors.

We can save the results to a file

```r
write.csv(q_f, file = "factors.csv",row.names=TRUE)
```
